```python
def day_cond(pdata, time_column_name='date_time', day=None, y=None, m=None, d=None):
    tmp1 = map(lambda x: str(x).split('-'), list(pdata[time_column_name]))
    tmp1 = list(tmp1)
    tmp = map(lambda x: x[2].split(' '), tmp1)
    tmp = list(tmp)
    if day != None:
        a, b, c = day.strftime('%Y-%m-%d').split('-')
    if (y!=None and m!=None and d!=None):
        a = y
        b = m
        c = d
    day_year  = str(a)
    day_month = str(b)
    day_day   = str(c)
    logic = np.logical_and(np.logical_and((np.array(tmp1)[:,0] == day_year), (np.array(tmp1)[:,1] == day_month)), (np.array(tmp)[:,0] == day_day))
    return logic
```

```python
def fetch_sensors_for_dates(data_pd, print_rate=-1):
    date_sensor_dict = {}
    prev_date = 'wat'
    n = len(data_pd)
    for i in range(n):
        curr_row  = data_pd.loc[i]
        if print_rate > 0:
            if i % print_rate == 0:
                print('current_iteration = ' ,i)
        curr_date = curr_row['timestamp'].split(' ')[0]
        if curr_date != prev_date:
            if prev_date != 'wat':
                date_sensor_dict[prev_date] = np.unique(date_sensor_dict[prev_date].split(',')[:-1])
            date_sensor_dict[curr_date] = str('')
        else:
            date_sensor_dict[curr_date] += curr_row['sensor'] + str(',')
        if i == n-1:
            date_sensor_dict[curr_date] = np.unique(date_sensor_dict[curr_date].split(',')[:-1])
        prev_date = curr_date
    return date_sensor_dict
```

```python
def check_sensors(date_sensor, sensors_list):
    suitable_dates = []
    for key in date_sensor.keys():
        date_sensor_list = list(date_sensor[key])
        intersection = set(date_sensor_list) & set(sensors_list)
        
        if len(intersection) == len(sensors_list):
            #print()
            suitable_dates.append(key)
    return suitable_dates
```

```python
def get_out_conditions_rus_foos(pdata):
    hours = []
    humidity = []
    pressure = []
    temp     = []
    wind_spd = []
    for i in data.index:
        hours.append(np.int64(data.loc[i]['   '].split(' ')[1].split(':')[0]) + np.int64(data.loc[i]['   '].split(' ')[1].split(':')[1]) / 60.)
        humidity.append(data.loc[i]['U'])
        pressure.append(data.loc[i]['P'] * 133.322)
        temp.append(data.loc[i]['T'])
        wind_spd.append(data.loc[i]['Ff'])
    
    hum   = interpolate.interp1d(hours, humidity, "cubic")
    press = interpolate.interp1d(hours, pressure, "cubic")
    t     = interpolate.interp1d(hours, temp, "cubic")
    w_sp  = interpolate.interp1d(hours, wind_spd, "cubic")
    return hum, press, t, w_sp
#out_hum, out_press, out_temp, out_wind_speed = get_out_conditions_foos(pdata, nextday)
```

```python
# MICROCLIMATE DYNAMICS FUNCTION OLD
def microclim(dtime, Tout, T, Noc, Co2, W, Q, get_diff=False):
    m = 540.     # air mass
    Cp = 1000.   # air heat capacity
    T_next = T + dtime/m/Cp*( 55.*(Tout-T) +
                              200.*(21.-T)  +
                              120.*Noc     +
                              W           +
                              Cp*Q*(21-T) +
                              0.1*m/3600.*Cp*(Tout-T))
    Co2_next = Co2 + dtime/m*(  (Q+0.1*m/3600.)*(300.-Co2) +
                                Noc*8. )
#    print(T_next, Co2_next)
    if get_diff==True:
        T_diff = dtime/m/Cp*( 55.*(Tout-T) +
                              200.*(21.-T)  +
                              120.*Noc     +
                              W           +
                              Cp*Q*(21-T) +
                              0.1*m/3600.*Cp*(Tout-T))
        Co2_diff = dtime/m*(  (Q+0.1*m/3600.)*(300.-Co2) +
                                Noc*8. )
        return T_diff, Co2_diff
    
    return T_next,Co2_next
```

```python
# MICROCLIMATE DYNAMICS FUNCTION
# setting the coefficients/constants
# room geometry - class A

height, area, width, F_window, F_nonwindow, delta_wall, ro_wall, lambda_wall, Cp_wall, m_wall, R_ogr = room('403')
alpha_int = 8.7 # from 
# M_metabolic = 120 # W/ per person - middle work
M_metabolic = 70 # W/ per person - low work
# human moisture transfer rate
M_moisture = 0.075 # kg/person hour
# human CO2 transfer rate
# M_co2 = 2 # ppm - middle work
M_co2 = 0.9 # ppm - low work
# number of lightning devices
N_light = 0.25*area
# heat power per lightning device
W_light = 20 # Wt
# notebook power
W_notebook = 100 # Wt
# the temperature difference between wall and room air
deltaT_wall = 3 # K from  = 4, but the calculation starts in almost steady-state
# air density inside the room
ro = 1.2754 # kg/m^3 
# gas constants
R_DryAir = 287.055 #J/kg*K 
R_wetvapor = 461.52 #J/kg*K 
# Heat of water evaporation
L_evap = 2501000 #J/kg 
# man with easy operation and comfortable temperature (20C) allocates 70 g of moisture per hour 
m_wateremmit = M_moisture/3600 #kg/s 
# Heat capacity
Cp_dry = 1004.8 #J/kgK dry air
Cp_vapor = 1.86 #J/kgK vapor
T_vent = 21 # ventilation temperature is constant (not dependent from ambient temperature)
m = ro*height*area # mass of wet air

def microclim_new(dtime, T_out, T, Noc, Co2, W, Q, phi, phi_out, V_wind, P_out, T_wall, m ):    
    # calculating the coefficients/constants
    
    # heat transfer coefficient for external enclosures ( )
    alpha_ext = 1.16*(5 + 10*V_wind**(1/2))
    P_satout = 611.2*math.exp(17.504*T_out/(241.2+T_out)) # saturated water vapor pressure in outside air
    P_satin = 611.2*math.exp(17.504*T/(241.2+T)) # saturated water vapor pressure in inside air
    
    k=1/(R_ogr+1/alpha_int) #thermal conductivity # Where 9 - total coefficient of heat transfer resistance of the outer wall
    
    P_in = P_out # take the pressure inside as outside
    
    #(T_out - T)*F_nonwindow/R_ogr # heat gain outside
       
    # functions of moisture
    d_out = R_DryAir/R_wetvapor*(phi_out*P_satout/(P_out-phi_out*P_satout)) #Moisture content of outdoor air
    d     = R_DryAir/R_wetvapor*(phi*P_satin/(P_out-phi*P_satin)) #Moisture content of indoor air
    Cp_in = Cp_dry + Cp_vapor*d   # air heat capacity
    Cp_out = Cp_dry + Cp_vapor*d_out # outside air heat capacity
    ro = P_out/(R_DryAir*(T+273.15)) - (d*P_out/(d+R_DryAir/R_wetvapor))*(1/R_DryAir-1/R_wetvapor)/((T+273.15))
    T_next = T + dtime/m/Cp_in*( width*height*k*(T_out-T)     + # cooling / heating of indoor air due to outside air flow (through slots)
                              # including wind speed
                              alpha_int*(area/width + width)*2*height*0.25*(T_wall-T) + # thermal inertia
                              M_metabolic*Noc                + # air heating due to human activity
                              W                              + # room heating / cooling with devices
                              Cp_out*Q*T_vent-T*Cp_in*Q        + # ventilation
                              0.2*m/3600.*Cp_out*(T_out-T)    - # infiltration
                              L_evap*M_moisture/3600.*Noc   + # moisture 
                              N_light*W_light                + # the effect of the light 
                              W_notebook*Noc*0)                  # equipment (computer*person)
                              
                                
    Co2_next = Co2 + dtime/m*(  (Q)*(300-Co2) + 0.2*m/3600*(300-Co2)+Noc*M_co2 )    
      
    T_wall_next = T_wall + dtime/m_wall/Cp_wall*(alpha_int*(area/width + width)*2*height*0.25*(T_next - T_wall)) # wall temperature
    
    d_next = d + dtime/m*(  (0.01*m/3600)*(d_out-d)+Q*(d_out) + Noc*M_moisture/3600.) 
    
    ro_next = P_out/(R_DryAir*(T_next+273.15)) - (d_next*P_out/(d_next+R_DryAir/R_wetvapor))*(1/R_DryAir-1/R_wetvapor)/(T_next+273.15)

    m = ro_next*height*area # mass of wet air 
    
    phi_next = d_next*P_out/((d_next+R_DryAir/R_wetvapor)*P_satin) 
    
    return T_next, Co2_next, phi_next, T_wall_next, m
```

```python
# FUNCTION FOR FLOW THROUGH VALVE (open area is not linear proportional to valve opening)
def valve_open(opening):
    xp = [0 , 1/9, 2/9, 3/9, 4/9, 5/9, 6/9, 7/9, 8/9, 9/9]
    fp = [0 , 1/144, 7/144, 16/144, 27/144, 43/144, 61/144, 84/144, 114/144, 144/144]
    f = interpolate.interp1d(xp, fp, "cubic")
    return f(opening)
```

```python
def charge_discharge(time): # charge = 1, discharge = -1
    xp = [0, 7.999, 8, 13.999, 14, 17.999, 18, 25]
    fp = [1,1,0,0,-1,-1,0,0]
    f = interpolate.interp1d(xp, fp, "linear")
    return f(time)

def chargetime(time): # charge = 1, discharge = -1
    xp = [0, 7.999, 8, 25]
    fp = [1,1,0,0]
    f = interpolate.interp1d(xp, fp, "linear")
    return f(time)

def dischargetime(time): # charge = 1, discharge = -1
    xp = [0, 13.999, 14, 17.999, 18, 25]
    fp = [0,0,-1,-1,0,0]
    f = interpolate.interp1d(xp, fp, "linear")
    return f(time)


# charge_discharge(0)
```

```python
# CONTROLLER FUNCTION TES
def OnOff_tes(time, T, T_prev, Co2,Co2_prev, phi, phi_prev,  Noc, heat_regime_prev,cool_regime_prev,vent_regime_prev, T_out, tes_type):

    # initialization
    if (T<23 and T>20):
        if (Noc > 20):
            heat_regime = 0
            cool_regime = 1
            vent_regime = 1
        else:
            heat_regime = 0
            cool_regime = 0
    else:
        heat_regime = heat_regime_prev
        cool_regime = cool_regime_prev

    if Co2 <750:
        vent_regime = 0
    else:
        vent_regime = vent_regime_prev
    
    if phi > 60:
        vent_regime = vent_regime_prev
    else:
        vent_regime = 0
    
    # controller
    if T>23:
        heat_regime = 0
        if (T>T_prev):
            if Noc > 20:
                cool_regime = min(cool_regime_prev+2,4)
            else:
                cool_regime = min(cool_regime_prev,4)

    if T < 20:
        cool_regime = 0
        if (T<T_prev and Noc < 15):
            heat_regime = min(heat_regime_prev+2,4)

    if Co2 > 750:
        if Co2>Co2_prev:
            if Noc > 20:
                vent_regime = min(vent_regime_prev+20,100)
            else:
                vent_regime = min(vent_regime_prev+20,100)
                
    if phi > 55:
        if phi>phi_prev:
            if Noc > 20:
                vent_regime = min(vent_regime_prev+20,100)
            else:
                vent_regime = min(vent_regime_prev+20,100)

    if Noc < 5:
        heat_regime = 0
        cool_regime = 0
        if T<18:
            heat_regime = 1
        if T>23:
            cool_regime = 1
    
    W_charge = 0
    W_charge_fortes = 50 # 3 kW*hours
    if charge_discharge(time) > 0:
        if tes_type == 'water':
            if T_water_melting - T_out > 4:
                W_charge = 0
            else:
                W_charge = W_charge_fortes
        if tes_type == 'ice':
            if T_ice_melting - T_out > 4:
                W_charge = 0
            else:
                W_charge = W_charge_fortes
        if tes_type == 'pcm1':
            if T_pcm_melting1 - T_out > 4:
                W_charge = 0
            else:
                W_charge = W_charge_fortes
        if tes_type == 'pcm2':
            if T_pcm_melting2 - T_out > 4:
                W_charge = 0
            else:
                W_charge = W_charge_fortes
        
    if charge_discharge(time) < 0:  
            cool_regime = 0
            vent_regime = max(60, vent_regime)


    return cool_regime,heat_regime,vent_regime, W_charge
```

```python
# MICROCLIMATE DYNAMICS FUNCTION WITH TES

def microclim_new_tes(dtime, T_out, T, Noc, Co2, W, Q, phi, phi_out, V_wind, P_out, T_wall, m, Q_tes):    
    # calculating the coefficients/constants
    
    # heat transfer coefficient for external enclosures ( )
    alpha_ext = 1.16*(5 + 10*V_wind**(1/2))
    P_satout = 611.2*math.exp(17.504*T_out/(241.2+T_out)) # saturated water vapor pressure in outside air
    P_satin = 611.2*math.exp(17.504*T/(241.2+T)) # saturated water vapor pressure in inside air
    
    k=1/(R_ogr+1/alpha_int) #thermal conductivity # Where 9 - total coefficient of heat transfer resistance of the outer wall
    
    P_in = P_out # take the pressure inside as outside
    
    #(T_out - T)*F_nonwindow/R_ogr # heat gain outside
       
    # functions of moisture
    d_out = R_DryAir/R_wetvapor*(phi_out*P_satout/(P_out-phi_out*P_satout)) #Moisture content of outdoor air
    d     = R_DryAir/R_wetvapor*(phi*P_satin/(P_out-phi*P_satin)) #Moisture content of indoor air
    Cp_in = Cp_dry + Cp_vapor*d   # air heat capacity
    Cp_out = Cp_dry + Cp_vapor*d_out # outside air heat capacity
    ro = P_out/(R_DryAir*(T+273.15)) - (d*P_out/(d+R_DryAir/R_wetvapor))*(1/R_DryAir-1/R_wetvapor)/((T+273.15))
        
    
    
    T_next = T + dtime/m/Cp_in*( width*height*k*(T_out-T)     + # cooling / heating of indoor air due to outside air flow (through slots)
                              # including wind speed
                              alpha_int*(area/width + width)*2*height*0.25*(T_wall-T) + # thermal inertia
                              M_metabolic*Noc                + # air heating due to human activity
                              W                              + # room heating / cooling with devices
                              Q_tes                           + # TES implemented
                              Cp_out*Q*T_vent-T*Cp_in*Q        + # ventilation
                              0.2*m/3600.*Cp_out*(T_out-T)    - # infiltration
                              L_evap*M_moisture/3600.*Noc   + # moisture 
                              N_light*W_light                + # the effect of the light 
                              W_notebook*Noc*0)                  # equipment (computer*person)
                              
                                
    Co2_next = Co2 + dtime/m*(  (Q)*(300-Co2) + 0.2*m/3600*(300-Co2)+Noc*M_co2 )    
      
    T_wall_next = T_wall + dtime/m_wall/Cp_wall*(alpha_int*(area/width + width)*2*height*0.25*(T_next - T_wall)) # wall temperature
    
    d_next = d + dtime/m*(  (0.01*m/3600)*(d_out-d)+Q*(d_out) + Noc*M_moisture/3600.) 
    
    ro_next = P_out/(R_DryAir*(T_next+273.15)) - (d_next*P_out/(d_next+R_DryAir/R_wetvapor))*(1/R_DryAir-1/R_wetvapor)/(T_next+273.15)

    m = ro_next*height*area # mass of wet air 
    
    phi_next = max(d_next*P_out/((d_next+R_DryAir/R_wetvapor)*P_satin),0) 
    
    return T_next, Co2_next, phi_next, T_wall_next, m, Cp_out
```

```python
# DIFFERENT TES DYNAMICS FUNCTION
# setting the coefficients/constants

# ICE

# Heat of ice melting
L_ice = 333000 #J/kg 

# Heat capacity of ice 
Cp_ice = 2100 #J/kg*K 
ro_ice = 900 #kg/m3
T_ice_melting = 0 #C


# WATER

# Heat capacity of water 
Cp_water = 4200 #J/kg*K 
ro_water = 1000 #kg/m3
T_water_melting = 0 #C

# PCM1 - K2HPO4(H2O)6

# Heat of pcm melting
L_pcm1 = 108000 #J/kg  PCM   

# Heat capacity of pcm 
Cp_pcm_fluid1 = 950 #J/kg*K  PCM   
Cp_pcm_solid1 = 950 #J/kg*K

# Density for pcm
ro_pcm1 = 1750 #kg/m3  PCM   

# Melting temperature
T_pcm_melting1 = 14 #C


# PCM2 - PCM Products Salt hydrate S13

# Heat of pcm melting
L_pcm2 = 160000 #J/kg  PCM   

# Heat capacity of pcm 
Cp_pcm_fluid2 = 1900 #J/kg*K  PCM   
Cp_pcm_solid2 = 1900 #J/kg*K

# Density for pcm
ro_pcm2 = 1515 #kg/m3  PCM   

# Melting temperature
T_pcm_melting2 = 13 #C

# TES VOLUMES

tes_height = 1 #m
tes_radius = 0.25 #m
V_total = tes_height*(tes_radius**2)*np.pi

# WATER/ICE

N_pipes_water = 4 # heat exchange pipes inside TES
d_pipe = 0.008 # m, external pipe diameter

V_dead_water = np.pi*(d_pipe**2)/4*(2*tes_height*N_pipes_water)

V_free_water = V_total - V_dead_water

m_water = ro_ice*V_free_water
m_ice = ro_ice*V_free_water

# PCM

N_pipes_pcm = 2*N_pipes_water # heat exchange pipes inside TES

V_dead_pcm = np.pi*(d_pipe**2)/4*(2*tes_radius*N_pipes_pcm)

V_free_pcm = V_total - V_dead_pcm

m_pcm1 = ro_pcm1*V_free_pcm
m_pcm2 = ro_pcm2*V_free_pcm


def tes(dtime, time, T_out, T, Q, Cp_out, air_mix_rate, charge, W_charge, tes_height, tes_radius, tes_type, T_tes,  T_tes_initial=16):    
    # calculating the coefficients/constants
    # PIPES INSIDE TES AND (FREE AND DEAD) VOLUME ARE DIFFERENT
    
#     #kostyl
#     T_out = 20
    
    W = W_charge
    if tes_type == 'water':
        if charge_discharge(time) > 0:
            if T_water_melting - T_out > 4:
                Q_tes = charge_discharge(time)*Cp_out*Q*(T_tes - T_out)
            else:
                Q_tes = charge_discharge(time)*W
        if charge_discharge(time) <= 0: 
            Q_tes = charge_discharge(time)*Cp_out*Q*(T_out-18) # 18 is air outtemp afet heat exchanger
        
        charge_next = charge + dtime*Q_tes
        
        if charge_next < m_water*Cp_water*(T_tes_initial - T_water_melting):
            T_tes_next = T_tes_initial - charge_next/m_water/Cp_water
        
        elif charge_next >= m_water*Cp_water*(T_tes_initial - T_water_melting):
            T_tes_next = T_water_melting
        
        if charge_next > m_water*Cp_water*(T_tes_initial - T_water_melting) + m_water*L_ice:
            diff = charge_next - m_water*Cp_water*(T_tes_initial - T_water_melting) + m_water*L_ice
            T_tes_next = T_water_melting - diff/m_water/Cp_ice
            
    if tes_type == 'ice':
        if charge_discharge(time) > 0: 
            if T_ice_melting - T_out > 4:
                Q_tes = charge_discharge(time)*Cp_out*Q*(T_tes - T_out)
            else:
                Q_tes = charge_discharge(time)*W
        if charge_discharge(time) <= 0:
            Q_tes = charge_discharge(time)*Cp_out*Q*(T_out-18)
        
        charge_next = charge + dtime*Q_tes
        
        if charge_next < m_ice*Cp_water*(T_tes_initial - T_ice_melting):
            T_tes_next = T_tes_initial - charge_next/m_ice/Cp_water
        
        elif charge_next >= m_ice*Cp_water*(T_tes_initial - T_ice_melting):
            T_tes_next = T_ice_melting
        
        if charge_next > m_water*Cp_water*(T_tes_initial - T_ice_melting) + m_ice*L_ice:
            diff = charge_next - m_ice*Cp_water*(T_tes_initial - T_ice_melting) + m_ice*L_ice
            T_tes_next = T_ice_melting - diff/m_ice/Cp_ice
            
    if tes_type == 'pcm1':
        if charge_discharge(time) > 0:
            if T_pcm_melting1 - T_out > 4:
                Q_tes = charge_discharge(time)*Cp_out*Q*(T_tes - T_out)
            else:
                Q_tes = charge_discharge(time)*W
        if charge_discharge(time) <= 0:
            Q_tes = charge_discharge(time)*Cp_out*Q*(T_out-18)
        
        charge_next = charge + dtime*Q_tes
        
        if charge_next < m_pcm1*Cp_pcm_fluid1*(T_tes_initial - T_pcm_melting1):
            T_tes_next = T_tes_initial - charge_next/m_pcm1/Cp_pcm_fluid1
        
        elif charge_next >= m_pcm1*Cp_pcm_fluid1*(T_tes_initial - T_pcm_melting1):
            T_tes_next = T_pcm_melting1
        
        if charge_next > m_pcm1*Cp_pcm_fluid1*(T_tes_initial - T_pcm_melting1) + m_pcm1*L_pcm1:
            diff = charge_next - m_pcm1*Cp_pcm_fluid1*(T_tes_initial - T_pcm_melting1) + m_pcm1*L_pcm1
            T_tes_next = T_pcm_melting1 - diff/m_pcm1/Cp_pcm_solid1
            
    if tes_type == 'pcm2':
        if charge_discharge(time) > 0:  
            if T_pcm_melting2 - T_out > 4:
                Q_tes = charge_discharge(time)*Cp_out*Q*(T_tes - T_out)
            else:
                Q_tes = charge_discharge(time)*W
        if charge_discharge(time) <= 0:
            Q_tes = charge_discharge(time)*Cp_out*Q*(T_out-18)
        
        charge_next = charge + dtime*Q_tes
        
        if charge_next < m_pcm2*Cp_pcm_fluid2*(T_tes_initial - T_pcm_melting2):
            T_tes_next = T_tes_initial - charge_next/m_pcm2/Cp_pcm_fluid2
        
        elif charge_next >= m_pcm2*Cp_pcm_fluid2*(T_tes_initial - T_pcm_melting2):
            T_tes_next = T_pcm_melting2
        
        if charge_next > m_pcm2*Cp_pcm_fluid2*(T_tes_initial - T_pcm_melting2) + m_pcm2*L_pcm2:
            diff = charge_next - m_pcm2*Cp_pcm_fluid2*(T_tes_initial - T_pcm_melting2) + m_pcm2*L_pcm2
            T_tes_next = T_pcm_melting2 - diff/m_pcm2/Cp_pcm_solid2
            
    if charge_next < 0:
        charge_next = 0
        T_tes_next = T_tes_initial
        if charge_next == 0:
            Q_tes = 0
        elif charge_next > 0:
            Q_tes = charge/dtime
    
    return charge_next, T_tes_next, Q_tes # J/s
```

```python
# NEW MODEL SIMULATION FUNCTION including TES
def RH_model_TES(TES='water', type_price='individual_3'):
    
    # classic
    
    stamp = []
    time = 0.1
    T_room = 20
    T_room_prev = 20
    phi_room = 50
    phi_room_prev = 50
    #print(phi_room)
    Noc = Noc_foo(time)
    Co2 = 400
    Co2_prev = 400
    HRP = 0
    CRP = 0
    VRP = 0
    Tout = outtemp_example(time, day='B')
    VENT = 100 # W - maximal power consumption by fan
    COP = 3 # coefficient of performance in regrigeration cycle
    
    dtime = 60 #in seconds
    energy_consumed = 0.0
    comfort_penalty = 0.0
    comfort_penalty_co2 = 0.0
    comfort_penalty_phi = 0.0
    phi_out = outhum(time)
    P_out = outpress(time)
    V_wind = outwindspeed(time)
    ro = 1.2754 # kg/m^3
    T_wall = T_room - deltaT_wall
    m = ro*height*area # mass of wet air
    m_prev = m
    T_wall_prev = T_wall
    
    # TES
    
    if Tout > 18:
        air_mix_rate = 0
    elif Tout > 15:
        air_mix_rate = 20
    else:
        air_mix_rate = 30
    T_tes_initial = 16
    charge = 0
    tes_type = TES
    Q_tes = 0
    Cp_out = Cp_dry
    T_tes_prev = T_tes_initial
    
    # COST
    
    bills = 0
    
    stamp.append([time, T_room, Noc, Co2, Tout, 0,0, phi_room, charge, T_tes_initial])
    
    while time<23.9:
        
        time += dtime/3600.
        Noc = Noc_foo(time)
        phi_out = outhum(time)
        
        Tout = outtemp_example(time, day='B')
        cool_regime,heat_regime,vent_regime, W_charge = OnOff_tes(time, T_room,T_room_prev, Co2,Co2_prev, phi_room, phi_room_prev,  Noc, HRP,CRP,VRP, Tout, tes_type)    
        HRP = heat_regime
        CRP = cool_regime
        VRP = vent_regime
        
        W = 400.*HRP-1500.*CRP
        Q = 0.55 / 100 * VRP
        
             
        T_room_prev = T_room
        Co2_prev = Co2
        phi_room_prev = phi_room
        T_wall_prev = T_wall
        m_prev = m
        T_tes_prev = T_tes_initial
        
        charge, T_tes, Q_tes = tes(dtime, time, Tout, T_room, Q, Cp_out, air_mix_rate, charge, W_charge, tes_height, tes_radius, tes_type, T_tes_prev, T_tes_initial=T_tes_initial)
        
        T_room, Co2, phi_room, T_wall, m, Cp_out = microclim_new_tes(dtime, Tout, T_room_prev, Noc, Co2_prev, W, Q, phi_room_prev, phi_out, V_wind, P_out, T_wall_prev, m_prev, Q_tes )
        
        # kostyl
        phi_room = 50
        
        
        stamp.append([time, T_room, Noc, Co2, Tout, W,Q, phi_room, charge, T_tes])

        energy_consumed += (abs(W) + (VRP/100)*VENT)*dtime/3600/1000
        
        bills += (abs(W/COP) + (VRP/100)*VENT)*dtime/3600/1000*price(time, type_price=type_price)
        
        if (T_room>24 and Noc>0):
            comfort_penalty += (T_room-24)*dtime/3600
        if (T_room<20 and Noc>0):
            comfort_penalty += (20-T_room)*dtime/3600
        if (Co2>800 and Noc>0):
            comfort_penalty_co2 += (Co2-800)*dtime/3600
        if (phi_room > 60 and Noc>0):
            comfort_penalty_phi += (phi_room-60)*dtime/3600
#         if (phi_room < 40 and Noc>0):
#             comfort_penalty_phi += (40-phi_room)*dtime/3600
        
    print("============================You consumed: ", "%0.0f" % energy_consumed, " kW*h per day" )
    print("=============================Daily bills: ", "%0.0f" % bills, " roubles" )
    print("======Level of discomfort by temperature: ", "%0.0f" % comfort_penalty, " Celsius*hour per day")
    print("Level of discomfort by CO2 concentration: ", "%0.0f" % comfort_penalty_co2, " ppm*hour per day")    
#     print("=========Level of discomfort by humidity: ", "%0.0f" % comfort_penalty_phi, " %*hour per day")
    return stamp

```
